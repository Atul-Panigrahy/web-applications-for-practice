```python
forr=[]
fromm=[]
date=[]
notice=[]

from urllib.parse import urljoin

from bs4 import BeautifulSoup
import requests

BASE_URL = 'https://hib.iiit-bh.ac.in/m-ums-2.0/start/login/'
with requests.Session() as session:
    response = session.get(BASE_URL)
    soup = BeautifulSoup(response.content, 'html.parser')
    for frame in soup.select("iframe"):
        frame_url = urljoin(BASE_URL, frame["src"])
        response = session.get(frame_url)
        frame_soup = BeautifulSoup(response.content, 'html.parser') 
        #print(frame_soup.prettify())a
        for framei in frame_soup.find_all('h5', class_='text-left'):
            cells=framei.find_all('font')
            #print(cells)
            if(len(cells)==1):
                notice.append(cells[0].find(text=True))
                
        for framei in frame_soup.find_all('footer', class_='text-left'):
            cells=framei.find_all('b')
            if(len(cells)==3):
                forr.append(cells[0].find(text=True))
                fromm.append(cells[1].find(text=True))
                date.append(cells[2].find(text=True))

import pandas as pd
df=pd.DataFrame(forr, columns=['target'])

#df['target']=forr
df['from']=fromm
df['date']=date
df['notice']=notice

df
```
